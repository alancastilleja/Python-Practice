```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = {'Male': 38, 'Female': 65, 'No': 36, 'Yes': 67, 'Male and Yes': 30, 'Male and No': 8, 'Female and Yes': 37, 'Female and No': 28}
data
```

```python
groups = list(data.keys())
values = list(data.values())
```

```python
fig = plt.figure(figsize = (25,10))
plt.rcParams['axes.titlesize'] = 18
ax1 = fig.add_subplot(141)
ax2 = fig.add_subplot(142)
ax3 = fig.add_subplot(143)
ax4 = fig.add_subplot(144)
ax1.title.set_text('Total people polled seperated by Gender')
ax1.set_ylabel('Number of People', fontsize = 18.0) 
ax1.set_xlabel('Gender', fontsize = 20) 
ax1.tick_params(axis = 'both', which = 'major', labelsize = 16)
ax2.title.set_text('Total votes of Yes and No')
ax2.set_ylabel('Number of People', fontsize = 18.0) 
ax2.set_xlabel('Choice', fontsize = 20)
ax2.tick_params(axis = 'both', which = 'major', labelsize = 16)
ax3.title.set_text('Yes and No for Males')
ax3.set_ylabel('Number of People', fontsize = 18.0) 
ax3.set_xlabel('Gender and Choice', fontsize = 20)
ax3.tick_params(axis = 'both', which = 'major', labelsize = 16)
ax4.title.set_text('Yes and No for Females')
ax4.set_ylabel('Number of People', fontsize = 18.0) 
ax4.set_xlabel('Gender and Choice', fontsize = 20)
ax4.tick_params(axis = 'both', which = 'major', labelsize = 16)
ax1 = ax1.bar(groups[0:2], values[0:2])
ax1[0].set_color('blue')
ax1[1].set_color('red')
ax2 = ax2.bar(groups[2:4],values[2:4])
ax2[0].set_color('royalblue')
ax2[1].set_color('orangered')
ax3 = ax3.bar(groups[4:6],values[4:6])
ax3[0].set_color('navy')
ax3[1].set_color('slategrey')
ax4 = ax4.bar(groups[6:8],values[6:8])
ax4[0].set_color('hotpink')
ax4[1].set_color('purple')
fig.suptitle('Does one\'s level of attractiveness make a difference in whether confidence is considered arrogance? n = 103', fontsize = 24)

plt.show()
```

According to the table above, it looks like attractiveness does make an impact especially amongst men. I would need numerical data such as age in order to do a regression to see if age and gender play a factor into saying yes or no.
